# Simple logistic model of iris

```python
df = iris.query("species == ('setosa', 'versicolor')") # like filter?
y_0 = pd.Categorical(df['species']).codes #make it 0/1 integer.
x_n = 'sepal_length' #just this character
x_0 = df[x_n].values #queries for values in this column. 
x_c = x_0 - x_0.mean() #centered. 
```

```python
# the model
with pm.Model() as model_0:
    α = pm.Normal('α', mu = 0, sd = 10)
    β = pm.Normal('β', mu = 0, sd = 10)
    μ = α + pm.math.dot(x_c, β) 
    θ = pm.Deterministic('θ', pm.math.sigmoid(μ)) #sigmoid of mu. 
    bd = pm.Deterministic('bd', -α/β) #criterion
    yl = pm.Bernoulli('yl', p = θ, observed = y_0)
    
    # trace
    trace_0 = pm.sample(1000)
```

```python
az.plot_trace(trace_0)
```

```python
theta = trace_0['θ'].mean(axis = 0)
idx = np.argsort(x_c)
plt.plot(x_c[idx], theta[idx], color = 'C2', lw = 3) 
plt.vlines(trace_0['bd'].mean(), 0, 1, color = 'k')
bd_hpd = az.hpd(trace_0['bd'])
plt.fill_betweenx([0, 1], bd_hpd[0], bd_hpd[1], color = 'k', alpha = 0.5)

plt.scatter(x_c, np.random.normal(y_0, 0.02),
           marker = '.', color = [f'C{x}' for x in y_0])
az.plot_hpd(x_c, trace_0['θ'], color = 'C2')

plt.xlabel(x_n)
plt.ylabel('θ', rotation = 0)

# original scale for xticks
locs, _ = plt.xticks()
plt.xticks(locs, np.round(locs + x_0.mean(), 1))
```

# Multiple logistic regression

```python
df = iris.query("species == ('setosa', 'versicolor')")
y_1 = pd.Categorical(df['species']).codes
x_n = ['sepal_length', 'sepal_width']
x_1 = df[x_n].values
```

```python
# model
with pm.Model() as model_1:
    α = pm.Normal('α', mu = 0, sd = 10)
    β = pm.Normal('β', mu = 0, sd = 2, shape = len(x_n)) #length 2. 
    μ = α + pm.math.dot(x_1, β)
    θ = pm.Deterministic('θ', 1 / (1 + pm.math.exp(-μ)))
    bd = pm.Deterministic('bd', -α/β[1] - β[0]/β[1] * x_1[:, 0])
    yl = pm.Bernoulli('yl', p = θ, observed = y_1)
    
    trace_1 = pm.sample(2000)
```

```python
# plot
idx = np.argsort(x_1[:, 0])
bd = trace_1['bd'].mean(0)[idx]
plt.scatter(x_1[:, 0], x_1[:, 1], c = [f'C{x}' for x in y_0])
plt.plot(x_1[:, 0][idx], bd, color = 'k')

az.plot_hpd(x_1[:, 0], trace_1['bd'], color = 'k')

plt.xlabel(x_n[0])
plt.ylabel(x_n[1])
```

# Softmax regression
one way to generalize logistic reg. to more than two classes.

```python
iris = sns.load_dataset('iris')
y_s = pd.Categorical(iris['species']).codes
x_n = iris.columns[:-1] 
x_s = iris[x_n].values
x_s = (x_s - x_s.mean(axis = 0)) / x_s.std(axis = 0)
```

```python
import theano.tensor as tt
```

```python
with pm.Model() as model_s: 
    α = pm.Normal('α', mu = 0, sd = 5, shape = 3) #three classes
    β = pm.Normal('β', mu = 0, sd = 5, shape = (4, 3)) #four predictors
    μ = pm.Deterministic('μ', α + pm.math.dot(x_s, β))
    θ = tt.nnet.softmax(μ)
    yl = pm.Categorical('yl', p = θ, observed = y_s)
    
    trace_s = pm.sample(2000)
```

```python
# how well did it predict? (.98)
data_pred = trace_s['μ'].mean(0)
y_pred = [np.exp(point)/np.sum(np.exp(point), axis = 0) for point in data_pred]
f'{np.sum(y_s == np.argmax(y_pred, axis = 1)) / len(y_s):.2f}'
```

```python
# model where we set some parameters to 0
with pm.Model() as model_sf:
    α = pm.Normal('α', mu = 0, sd = 2, shape = 2)
    β = pm.Normal('β', mu = 0, sd = 2, shape = (4, 2))
    α_f = tt.concatenate([[0], α])
    β_f = tt.concatenate([np.zeros((4, 1)), β], axis = 1)
    μ = α_f + pm.math.dot(x_s, β_f)
    θ = tt.nnet.softmax(μ)
    yl = pm.Categorical('yl', p = θ, observed = y_s)
    
    #trace
    trace_sf = pm.sample(1000)
```

```python
soft1 = az.summary(trace_s, var_names = ['α', 'β']) #high sd.
soft2 = az.summary(trace_sf, var_names = ['α', 'β']) #low sd. 
```

# Poisson regression
for count data.

number of events in fixed interval of time and/or space. 

only one parameter (rate: $\mu$): both mean and variance.

$f(x \mid y) = \frac{e^{-\mu}\mu^{x}}{x!}$, where: 

$\mu$ avg. n. of events per unit of time/space. <br/>
$x$ positive integer values ${0, 1, \ldots, n}$ <br/>
$x!$ factorial of $x$.

```python
from scipy import stats
```

```python
# visualizing Poisson
mu_params = [0.5, 1.5, 3, 8]
x = np.arange(0, max(mu_params) * 3)
for mu in mu_params:
    y = stats.poisson(mu).pmf(x)
    plt.plot(x, y, 'o-', label = f'μ = {mu:3.1f}')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
```

# Zero inflated Poisson
mixture of two models that generate posterior.
(1) Poisson.
(2) just generates zeros.


```python
n = 100
θ_real = 2.5
ψ = 0.1

# simulate 
counts = np.array([(np.random.random() > (1 - ψ)) * 
                  np.random.poisson(θ_real) for i in range(n)])
```

```python
# built in ZIP distribution from PyMC3
# do this manually if it is not in tasks
with pm.Model() as ZIP:
    
    # priors 
    ψ = pm.Beta('ψ', 1, 1)
    θ = pm.Gamma('θ', 2, 0.1)
    
    # likelihood 
    y = pm.ZeroInflatedPoisson('y', ψ, θ, observed = counts)
    
    # trace
    trace = pm.sample(1000)
```

```python
az.plot_trace(trace)
```

# Robust logistic regression

```python
iris = sns.load_dataset("iris")
df = iris.query("species == ('setosa', 'versicolor')")
y_0 = pd.Categorical(df['species']).codes
x_n = 'sepal_length'
x_0 = df[x_n].values
y_0 = np.concatenate((y_0, np.ones(6, dtype = int))) #add intruders
x_0 = np.concatenate((x_0, [4.2, 4.5, 4.0, 4.3, 4.2, 4.4])) #add intruders
x_c = x_0 - x_0.mean()
plt.plot(x_c, y_0, 'o', color = 'k');
```

```python
# model
with pm.Model() as model_rlg:
    α = pm.Normal('α', mu = 0, sd = 10)
    β = pm.Normal('β', mu = 0, sd = 10)
    μ = α + x_c * β
    θ = pm.Deterministic('θ', pm.math.sigmoid(μ))
    bd = pm.Deterministic('bd', -α/β)
    π = pm.Beta('π', 1., 1.)
    p = π * 0.5 + (1 - π) * θ
    yl = pm.Bernoulli('yl', p = p, observed = y_0)
    
    trace_rlg = pm.sample(1000)
```

```python
# now for our own model (petal length/width instead of sepal). 
# preprocess data. 

df = iris.query("species == ('setosa', 'versicolor')") # like filter?
y_0 = pd.Categorical(df['species']).codes #make it 0/1 integer.
x_length = 'petal_length' #just this character
x_width = 'petal_width'
x_0l = df[x_length].values #queries for values in this column. 
x_0w = df[x_width].values
x_cl = x_0l - x_0l.mean() #centered. 
x_cw = x_0w - x_0w.mean()
```

```python
# make function
def plot_hdi(alpha, interval, trace, x_centered, x_name, x_origvalues, y_outcome):
    
    theta = trace['θ'].mean(axis = 0)
    idx = np.argsort(x_centered)
    plt.plot(x_centered[idx], theta[idx], color = 'C2', lw = 3)
    plt.vlines(trace['bd'].mean(), 0, 1, color = 'k')
    bd_hdi = az.hdi(trace['bd'], interval)
    plt.fill_betweenx([0, 1], bd_hdi[0], bd_hdi[1], color = 'k', alpha = alpha)
    
    plt.scatter(x_centered, np.random.normal(y_outcome, 0.02),
               marker = '.', color = [f'C{x}' for x in y_outcome])
    az.plot_hpd(x_centered, trace['θ'], color = 'C2')
    
    plt.xlabel(x_name)
    plt.ylabel('θ', rotation = 0)
    
    locs, _ = plt.xticks()
    plt.xticks(locs, np.round(locs + x_origvalues.mean(), 1))
    
```

```python
# the length model
plot_hdi(alpha = 0.5, interval = .95, trace = e1_length_t, 
        x_centered = x_cl, x_name = x_length, x_origvalues = x_0l,
        y_outcome = y_0)
```

```python
# the width model
plot_hdi(alpha = 0.5, interval = .95, trace = e1_width_t, 
        x_centered = x_cw, x_name = x_width, x_origvalues = x_0w,
        y_outcome = y_0)
```

```python
# the model
with pm.Model() as E3_lin:
    
    # priors 
    α = pm.Normal('α', mu = 0, sd = 10)
    β = pm.Normal('β', mu = 0, sd = 10)
    ϵ = pm.HalfCauchy('ϵ', 5)
    
    # linear model 
    μ = pm.Deterministic('μ', α + x_c * β)
    #μ = α + pm.math.dot(x_c, β)
    
    # likelihood
    y = pm.Normal('y', mu = μ, sigma = ϵ, observed = y_0)
```

```python
# gathering stuff
with E3_lin:
    
    # trace
    E3_lin_t = pm.sample(1000)
    
    # posterior predictive 
    E3_pp = pm.sample_posterior_predictive(E3_lin_t)
    print(az.summary(E3_lin_t, credible_interval=.95))
```

# Exercise 5
Use the same example from the previous exercise. For model_1, check how
much the log-odds change when increasing sepal_length from 5.5 to 6.5
(spoiler: it should be 4.66). How much does the probability change? How does
this increase compare to when we increase from 4.5 to 5.5?

```python
# compute log-odds
def prob_fun(sep_length, sep_width, df, diff):
    
    x_1 = sep_length
    x_2 = sep_width
    
    log_odds_versicolor_i = (df['mean'] * [1, x_1, x_2]).sum()
    probability_versicolor_i = logistic(log_odds_versicolor_i)
    
    log_odds_versicolor_f = (df['mean'] * [1, x_1 + diff, x_2]).sum()
    probability_versicolor_f = logistic(log_odds_versicolor_f)
    
    log_diff = log_odds_versicolor_f - log_odds_versicolor_i
    prob_diff = probability_versicolor_f - probability_versicolor_i
    
    return(log_diff, prob_diff)
```

```python
for sepal_length in [4.5, 5.5]: 
    diff = 1
    log_diff, prob_diff = prob_fun(sep_length = sepal_length,
                                   sep_width = 3, 
                                   df = val, 
                                   diff = diff)
    print(f'sepal i = {sepal_length}, sepal f = {sepal_length + diff} \nlog difference: {log_diff}, prob. diff: {prob_diff}')
```

## from his:
"From the calculation above we see that while the log-odds change stays constant, as it should in linear regression, the probability change is not as large from 5.5 to 6.5 as it is from 4.5 to 5.5. Looking at the graphic this intuitively makes sense as well. When sepal length is at 4.5, the chance that the species is versicolor is very small. When sepal length jumps to 5.5, this probability gets a lot bigger. This means that subsequently going from 5.5 to 6.5 still increases the probability of versicolor, but not as much - because, well, at 5.5 there is already a good chance that the species we're seeing is versicolor."

```python
# https://docs.pymc.io/notebooks/GLM-robust.html

# create toy data
size = 100
true_int = 1
true_slope = 2

x = np.linspace(0, 1, size)
true_reg = true_int + true_slope * x
y = true_reg + np.random.normal(scale = 0.5, size = size)

x_out = np.append(x, [0.1, 0.15, 0.2])
y_out = np.append(y, [8, 6, 9])

data = dict(x = x_out, y = y_out)
```

```python
# estimated standardly
with pm.Model() as model:
    pm.glm.GLM.from_formula("y ~ x", data)
    trace = pm.sample(2000, cores=2)
```

```python
# results in skewed fit
plt.figure(figsize=(7, 5))
plt.plot(x_out, y_out, "x", label="data")
pm.plot_posterior_predictive_glm(trace, samples=100, label="posterior predictive regression lines")
plt.plot(x, true_reg, label="true regression line", lw=3.0, c="y")

plt.legend(loc=0);
```
