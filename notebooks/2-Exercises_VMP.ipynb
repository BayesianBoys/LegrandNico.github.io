For this exercise we will generate 57 datapoints from a distribution of $Normal(4, 0.5)$:

```python
data = np.random.normal(loc = 4, scale = 0.5, size = 200)

with pm.Model() as benko_gambit: 
    
    mu = pm.Normal("mu", 0, 10)
    sigma = pm.HalfNormal("sigma", 25)
    likelihood = pm.Normal("likelihood", mu = mu, sigma = sigma, observed = data)
    
    trace = pm.sample(draws = 5000)
    post_checks = pm.sample_posterior_predictive(trace, samples = 200, random_seed = 42)
    prior_checks = pm.sample_prior_predictive(samples = 200, random_seed= 42)
```

```python
# very bad prior. 
prior_preddy = az.from_pymc3(trace = trace, posterior_predictive = prior_checks) 
az.plot_ppc(prior_preddy, figsize=(12, 6), mean=False)
```

```python
howell = pd.read_csv("https://raw.githubusercontent.com/LegrandNico/CognitiveModeling/master/notebooks/data/howell.csv", delimiter=";")
```

```python
# only grown up pleebs. 
h_sub = howell[howell["age"] > 18]
```

```python
h_sub['height'].values
```

```python
# make da model
with pm.Model() as h_mod: 
    
    # priors
    alpha = pm.Normal("alpha", 0, 10)
    beta = pm.Normal("beta", 0, 1)
    epsilon = pm.HalfNormal("epsilon", 10) 
    
    # linear model
    lin_model = pm.Deterministic("lin_model", alpha + beta * h_sub['weight'].values)

    # likelihood -> posterior
    outcome = pm.Normal("likelihood", mu = lin_model, sd = epsilon, observed = h_sub['height'].values)
```

```python
with h_mod:
    trace = pm.sample(1000, tune = 2000)
    prior_pred = pm.sample_prior_predictive(trace, 100) # does not work.
    post_pred = pm.sample_posterior_predictive(trace, 100)
```

```python
az.plot_trace(trace, var_names = ['alpha', 'beta', 'epsilon'])
```

```python
# prior predictive 
prior_preddy = az.from_pymc3(trace = trace, posterior_predictive = prior_pred) 
az.plot_ppc(prior_preddy, figsize=(12, 6), mean=False)
```

```python
# posterior predictive 
post_preddy = az.from_pymc3(trace = trace, posterior_predictive = post_pred) 
az.plot_ppc(post_preddy, figsize=(12, 6), mean=False)
```
